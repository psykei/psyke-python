```python
import numpy as np
from psyke import Extractor
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import metrics
import plotly.express as px
import math


df_path="datasets\df.txt"
```

```python
df=pd.read_csv(df_path,sep=',')
```

```python
df.head()
```

```python
target_list=['Z0','Z1','Z2','Z3','Z4','Z5','Z6','Z7','Z8','Z9']
select_var_Z=5
select_name_Z='Z5'
```

```python
X=df[['X','Y']]
Y=df[target_list]
df=pd.DataFrame(pd.concat([X,Y],axis=1))
df=df.dropna().reset_index(drop=True)
```

```python
X=df[['X','Y']]
Y=df[target_list]
```

```python
class_expected=np.unique(Y)
print("Number of expected class are:", class_expected)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
X_train=X_train.to_numpy()
X_test=X_test.to_numpy()
Y_train=Y_train.to_numpy()
Y_test_testing=Y_test
Y_test=Y_test.to_numpy()
```

```python

Y_train_Z=[]
for i in range(len(Y_train)):
    Y_train_Z.append(Y_train[i][select_var_Z])
```

```python

Y_test_Z=[]
for i in range(len(Y_test)):
    Y_test_Z.append(Y_test[i][select_var_Z])
```

```python
#add seaborn pairplot
import matplotlib.pyplot as plt
fig = px.scatter_matrix(X,
    dimensions=['X','Y'])
fig.show()
```

We assume to search for only 2 clusters

```python
Y_train_Z=[math.floor(x) for x in Y_train_Z]
```

```python

Y_test_Z=[]
for i in range(len(Y_test)):
    Y_test_Z.append(Y_test[i][select_var_Z])
```

```python
exact_cluster_to_search=len(np.unique(Y_train_Z))
n_cluster=4 
```

```python
exact_cluster_to_search
```

```python
#initialize with n clusters
imm = Extractor.imm(exact_cluster_to_search)
imm.fit(X_train)
imm.drawTree()

Y_pred_1=imm.predict_on_more_data(X_test)

print('rand_score is: ',metrics.rand_score(Y_test_Z,Y_pred_1),' , ARI score is : ',metrics.adjusted_rand_score(Y_test_Z,Y_pred_1),' , AMI score is: ',metrics.adjusted_mutual_info_score(Y_test_Z,Y_pred_1),' , silhoutte score is: ',metrics.silhouette_score(X_test,Y_pred_1))  
```

```python
#initialize with n clusters
imm = Extractor.imm(n_cluster)
imm.fit(X_train)
imm.drawTree()

Y_pred_2=imm.predict_on_more_data(X_test)

print('rand_score is: ',metrics.rand_score(Y_test_Z,Y_pred_2),' , ARI score is : ',metrics.adjusted_rand_score(Y_test_Z,Y_pred_2),' , AMI score is: ',metrics.adjusted_mutual_info_score(Y_test_Z,Y_pred_2),' , silhoutte score is: ',metrics.silhouette_score(X_test,Y_pred_2))  
```
